```python
subway = pd.read_csv('./data/서울교통공사_역주소 및 전화번호.csv')
```

```python
subway
```

```python
subway['구'] = subway['역주소'].str.split(" ").str[1]
```

```python
subway = subway.drop(['연번','역번호','호선','역전화번호'], axis = 1)
subway
```

```python
subway.isnull().sum()
```

```python
 subway['역명'] =subway['역명'].apply(lambda x: x if x[-1]=='역' else x + '역')
```

```python
subway
```

```python
subway = subway.drop_duplicates(['역명'])
```

```python
subway
```

```python

for i in range(len(subway)):
    try:
        subway['역주소'][i] = re.sub("\(.*\)|\s-\s.*", '', subway['역주소'][i])
        print(subway['역주소'][i])
    except:
        pass
```

```python
subway
```

```python
sub_name = []
for i in subway['역주소']:
    sub_name.append(i)
```

```python
len(sub_name)
```

```python
gmaps = googlemaps.Client('개인 api')
lat = []
lon = []
for i in sub_name:
    
    location = gmaps.geocode(i, language='ko') 
    
    latitude  = location[0]["geometry"]["location"]["lat"] 
    longitude = location[0]["geometry"]["location"]["lng"] 

    lat.append(latitude)
    lon.append(longitude)
    print(i)
```

```python
subway['위도'] = lat
subway['경도'] = lon
subway
```

```python
subway.to_csv('./data/subway.csv', header = True, index = False)
```
