```python
# 서울시 소방서 데이터 추가
fire_station = pd.read_csv('./data/서울시 소방서,안전센터,구조대 위치정보.csv', engine='python')
fire_station
```

```python
# WGS84 : x,y좌표 
# GRS80 : 경도, 위도
!pip install pyproj
```

```python
from pyproj import Proj, transform

# Projection 정의
# UTM-K
proj_UTMK = Proj(init='epsg:5186') # 중부원점일때

# WGS1984
proj_WGS84 = Proj(init='epsg:4326') # Wgs84 경도/위도, GPS사용 전지구 좌표
```

```python
lat = []
lng = []
for i in range(len(fire_station)):
    x = fire_station['X좌표'][i]
    y = fire_station['Y좌표'][i]
    lat_x, lng_y = transform(proj_UTMK,proj_WGS84,x,y)
    lat.append(lat_x)
    lng.append(lng_y)
```

```python
fire_station.columns
```

```python
fire_station = fire_station.drop(['연번', '서ㆍ센터ID','도서지역포함여부', 
                                  '상위서ㆍ센터ID', '일련번호', 
                                  'X좌표','Y좌표'], axis=1)
```

```python
fire_station['lng'] = lat
fire_station['lat'] = lng
fire_station
```

```python
fire_station = fire_station[fire_station['유형구분명'] == '소방서']
```
